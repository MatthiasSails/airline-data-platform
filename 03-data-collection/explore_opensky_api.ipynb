# OpenSky Network API — Exploration
Testing the OpenSky client: mock mode and real API.

```python
import sys
sys.path.append('.')

from opensky_api.client import OpenSkyClient

client = OpenSkyClient(use_mock=True)
print('Client erstellt (Mock-Modus)')
```

## 1. Mock mode — departures Berlin Brandenburg (EDDB)

```python
departures = client.get_departures(airport='EDDB', begin=0, end=1)

print(f'Departures: {len(departures)}')

for f in departures[:3]:
    cs = (f.get('callsign') or '').strip()
    print(f"  {cs} | {f['estDepartureAirport']} → {f['estArrivalAirport']}")
```

## 2. Mock mode — arrivals Berlin Brandenburg (EDDB)

```python
arrivals = client.get_arrivals(airport='EDDB', begin=0, end=1)

print(f'Arrivals: {len(arrivals)}')
for f in arrivals:
    print(f"  {f['callsign'].strip()} | {f['estDepartureAirport']} → {f['estArrivalAirport']}")
```

## 3. Live API — departures from the last 24 hours

```python
import time

real_client = OpenSkyClient(use_mock=False)

end   = int(time.time())
begin = end - 60 * 60 * 24  # 24 hours back

real_departures = real_client.get_departures(airport='EDDB', begin=begin, end=end)
print(f'Live departures from EDDB (last 24h): {len(real_departures)}')

for f in real_departures[:5]:
    cs = (f.get('callsign') or '').strip()
    print(f"  {cs} | {f['estDepartureAirport']} → {f['estArrivalAirport']}")
```
